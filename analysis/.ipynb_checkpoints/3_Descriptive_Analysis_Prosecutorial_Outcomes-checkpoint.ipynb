```python
import pandas as pd
import os
import sys
sys.path.append('..')

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

lab_blue = '#2b4888'
lab_pink = '#de4057'
lab_grey = '#595959'

DATA_DIR = os.path.join('..', 'data')
```

# Data Munging

```python
# Read in data

# prosecutorial outcomes from USAO
usao = pd.read_csv(os.path.join(DATA_DIR, 'USAO_data.csv'), 
                               dtype = {'CCN': str},                               
                               parse_dates = ['Arrest Date', 'Defendant DOB'])                               

# NIBIN information from the Department of Forensic Sciences 
nibin_hits = pd.read_excel(os.path.join(DATA_DIR, 
                                        'NIBIN_Entry_thelab_2017-2019.xlsx'), 
                                        sheet_name='nibin_hits')

# This pulls in a clean copy of the data, which comes from the notebook: 
# 4_Quasi_Experimental_Analysis_Synthetic_Control.ipynb (This notebook was originally done first, before reorganization)
dcr_crime = pd.read_csv(os.path.join(DATA_DIR, 'crime_df_usao.csv'), 
                        dtype = {'CCN':str, 'Offense Location PSA': str})

# Original arrest file from Daniel
arrests = pd.read_csv(os.path.join(DATA_DIR, 'AllArrests_171101_190430.csv'), 
                        dtype = {'CCN':str}, encoding = 'cp1252', parse_dates = ['Arrest Date'])
```

```python
# Clean NIBIN information and merge to USAO prosecutorial outcomes

nibin_hits['CCN'] = nibin_hits['CCN'].str.replace('-', '') #Remove hyphens
nibin_hits['CCN'] = nibin_hits['CCN'].str[-8:] #Get the last 8 digits

# Merge hit information to USAO prosecutorial outcomes, to see whether they had hits
usao_hits = usao.merge(nibin_hits[['CCN', 'Hit Count']], 
                       how = 'left', 
                       on = 'CCN')

# Merge all info to arrests data to get the arrest district; 
usao_hits = usao_hits.merge(arrests[['CCN', 'Arrest Number#','Arrest Location District']], 
                            how = 'left', 
                            on = ['CCN','Arrest Number#']) 

usao_hits.drop_duplicates(inplace=True)

# Merge to DCR crimes data to get Offense Category
usao_hits = usao_hits.merge(dcr_crime[['CCN', 'Offense Category']], 
                            how = 'left', 
                            on = 'CCN')

# Now clean up the dataframe
usao_hits.rename(columns = {'CCN' : 'ccn', 
                            'Arrest Number#' : 'arrest_id', 
                            'Arrest Date': 'event_date',
                            'Arrest Investigation Category/Description': 'category',
                            'Defendant PDID' : 'pdid', 
                            'Defendant DOB' : 'dob',
                            'Prosecution Status (as of July 2019)' : 'status',
                            'Next Court Date' : 'next_appearance',
                            'District Court' : 'district_court', 
                            'Offense Category' : 'offense_category', 
                            'Arrest Location District' : 'arrest_district'},
                 inplace = True)


# Create a flag for whether there was a hit
usao_hits['hit'] = usao_hits['Hit Count']>0

usao_hits.drop(columns=(['Defendant Last Name', 
                         'Defendant First Name', 
                         'Hit Count']), inplace = True)

usao_hits.drop_duplicates(inplace=True)

# Create a flag if in 7D
usao_hits['sevenD'] = usao_hits.arrest_district == '7D'

usao_hits.shape
```

```python
#Clean the "status" column of our dataset
usao_hits['status'] = usao_hits.status.str.strip()
usao_hits['status'] = usao_hits.status.str.title()

#Grand jury no bill means the USAO took the case, but the suspect was not indicted so the case was dismissed.
fix_spellings = {'Oag': 'OAG',
                 'Oag Charge': 'OAG',
                 'Unable To Locate This Ccn' : 'Unable To Locate',
                 'Unable Ot Locate' : 'Unable To Locate',
                 'This Defendant Not On The Ccn???' : 'Unable To Locate',
                 'No Papered': 'No Paper', 
                 'Paperd' : 'Papered',
                 'Papering': 'Papered', 
                 'Papered; Still Pending':'Papered',
                 'Papered; Still Pending; Tral' : 'Papered',
                 'Grand Jury No Bill': 'Dismissed', 
                 'No Papered, Pending Dna Results': 'Not Papered, Pending DNA Results'}

usao_hits.status.replace(fix_spellings, inplace=True)
```

```python
# There were quite a few pdids which did not start with 'DC'. To make sure I don't have duplicates, I am:
# grabbing the non-null ids (bc otherwise I can't filter properly) - there are 26 missing pdids
# pulling out the ids that don't start with 'DC', then adding 'DC' to the string'
# adding those back to the list, dropping straight up duplicates, and appending it back to the nulls


not_null_pdid = usao_hits[~usao_hits.pdid.isna()] #Get the non-null pdids

#pull out the ones that did not have a 'DC' at the beginning, and add 'DC' to the beginning
not_startDC_pdid = not_null_pdid[~not_null_pdid.pdid.str.startswith('DC')].copy() 
not_startDC_pdid['pdid'] = 'DC' + not_startDC_pdid.pdid

#Append the fixed pdids to the non-null-pdids that DID start with 'DC'
#This is now all of the cleaned pdids
not_null_pdid = not_null_pdid[not_null_pdid.pdid.str.startswith('DC')].append(not_startDC_pdid)
not_null_pdid.drop_duplicates(inplace = True)

#Append the cleaned pdids to the df with the missing pdid
usao_hits = usao_hits[usao_hits.pdid.isna()].append(not_null_pdid)
```

```python
# Printing out information
print('How many unique arrests?', usao_hits.arrest_id.nunique())
print('Across how many cases during our time frame?', usao_hits.ccn.nunique())
print('How many unique defendants?', usao_hits.pdid.nunique())
```

```python
def generateCrosstabs(df):
    '''
    Function to generate crosstabs specifically for USAO data
    This takes the data and counts the number and percentage of 
    hits and no hits (or no ballistics evidence) by prosecutorial status
    '''
    
    xtab = pd.crosstab(df.status, df.hit, 
                       margins = True, margins_name="Total")\
                       .reset_index()\
                       .sort_values(by = 'Total', ascending = False)[1:]
    xtab.rename(columns = {False:'Ballistics Evidence Only', 
                           True:'NIBIN Hit'}, inplace = True)
    
    #Calculate percentages
    xtab['Hit'] = round(xtab['NIBIN Hit']/xtab['NIBIN Hit'].sum()*100,2)
    xtab['No Hit'] = round(xtab['Ballistics Evidence Only']/xtab['Ballistics Evidence Only'].sum()*100,2)
    
    return xtab
```

```python
def showDataLabels(df, varname):
    ''' 
    Function for creating data labels
    '''
    # Get current axis on current figure
    ax = plt.gca()
    
    n = list((df[varname]).astype(int)) #Raw value
    listerator = 0 #my iterator for the list

    for p in ax.patches: 
        height = p.get_height()
        width = p.get_x() + p.get_width()/2.
        height = np.where(np.isnan(height), 0, height)
        ax.text(x = width, 
                y = height, 
                s = '%.1f' % float(height) + '%\n' + 'N=' + str(n[listerator]),
                fontsize=11, 
                color='black', 
                ha='center', 
                va='bottom')
        if listerator < len(n)-1:
            listerator += 1
```

```python
known_list = ["Guilty Plea", "Dismissed", "No Paper", "No Paper (Co-Def Was Charged)",
              "Trial Guilty", "Not Guilty At Trial", "Pled To Other Charges"]

unknown_list = ["Papered", "Not Papered, Pending DNA Results", "N/A (Juvenile)", 
                "Unable To Locate", "OAG", "Defendant Deceased"]

#Filter so that the statuses line up to the list above. 
known_outcomes = usao_hits[usao_hits.status.isin(known_list)]
unknown_outcomes = usao_hits[usao_hits.status.isin(unknown_list)]

known_outcomes = generateCrosstabs(known_outcomes)
unknown_outcomes = generateCrosstabs(unknown_outcomes)
```

```python
# ---------------- Known Outcomes -------------- #
# Reshape the dataframe from wide to long for plotting putposes
known = pd.melt(known_outcomes, id_vars='status', value_vars = ['Ballistics Evidence Only', 'NIBIN Hit'])

# Get the total values so we can calculate percentages
ballistics_only = known[known.hit == 'Ballistics Evidence Only'].sum()['value']
hit_only = known[known.hit == 'NIBIN Hit'].sum()['value']

# Create a new column w the percentages
known['pct'] = np.where(known.hit == 'Ballistics Evidence Only', 
                        known.value/ballistics_only*100, 
                        known.value/hit_only*100)

# --------------- Unknown Outcomes ------------- #
unknown = pd.melt(unknown_outcomes, id_vars='status', value_vars = ['Ballistics Evidence Only', 'NIBIN Hit'])
# Get the total values so we can calculate percentages
ballistics_only = unknown[unknown.hit == 'Ballistics Evidence Only'].sum()['value']
hit_only = unknown[unknown.hit == 'NIBIN Hit'].sum()['value']

# Create a new column w the percentages
unknown['pct'] = np.where(unknown.hit == 'Ballistics Evidence Only', 
                        unknown.value/ballistics_only*100, 
                        unknown.value/hit_only*100)
```

##  Outcomes for all of DC

```python
# For known prosecutorial outcomes

#Plot the count of statuses/outcomes by whether the case had a hit or no hit. 
plt.figure(figsize=(18,10))
sns.barplot(x = 'status', y = 'pct', data = known, hue = 'hit', 
            palette = [lab_blue, lab_pink])

plt.ylabel('Percentage of Outcomes', fontsize = 12)
plt.xlabel('', fontsize = 12)
plt.title('Prosecutorial Outcomes for Arrests in Cases with any Ballistics Evidence in all of DC\n'+
          '(Cases with known outcomes)', fontsize = 14)

plt.ylim(0,90)
plt.legend()

showDataLabels(df = known, varname = 'value')
```

```python
# For unknown prosecutorial outcomes

plt.figure(figsize=(20,10))
sns.barplot(x = 'status', y = 'pct', data = unknown,
            hue = 'hit', palette = [lab_blue, lab_pink])

plt.ylabel('Percentage of Outcomes', fontsize = 12)
plt.xlabel('', fontsize = 12)
plt.title('Prosecutorial Outcomes for Arrests in Cases with any Ballistics Evidence in all of DC\n'+
          '(Cases with Unknown Outcomes)', fontsize = 14)

plt.ylim(0,90)
plt.legend()

showDataLabels(df=unknown, varname='value')
```
